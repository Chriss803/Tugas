```python
#Importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line plots
x,y =[5,10,15,20,25],[5,10,15,20,25]
plt.plot(x,y)
plt.show
```

```python
plt.plot([1,5,9],[1.4,4,3.2],ls=':',color="red")
plt.plot([9,8,1],[1.4,3,5.6],ls=':',color="blue")
plt.show()
```

```python
plt.plot([1,2,3,4,5,6,7,8],[8,7,6,5,4,3,2,1],'k_-',label="Black Line")
plt.plot([1,2,3],[3,2,1],'r_-',label="Red Line")
# Show Legend
plt.legend(fontsize=15)
plt.xlim(1,10) #Setting X-Axis limits
plt.ylim(1,10) #Setting Y-Axis limits
# plt.show()
(1.0, 10.0)
```

```python
# importing math
import math
# Creating some random data
X = np.arange(0, math.pi*2, 0.01)
Y1 = np.sin(X)
Y2 = np.cos(X)
# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(nrows=1, ncols=2,figsize=(13,5))
# For Sine Function
axis[0].plot(X, Y1)
axis[0].set_title("Sine Function")
# For Cosine Function
axis[1].plot(X, Y2)
axis[1].set_title("Cosine Function")
plt.show()
```
